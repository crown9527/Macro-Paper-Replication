```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def investment_ratio(ϕ,γ,r):
    if (ϕ > -10):
        ratio=(γ/((1+r)*(1-γ)))**(1/(1-ϕ))
    else:
        ratio= 1 
    return ratio

phi_total=np.array([-100,-5,-0.5,0,0.5,0.9])
phi_size=np.size(phi_total)
gamma_total =np.linspace(0.1,0.9,100)
gamma_size=np.size(gamma_total)
invest_ratio=np.zeros([6,100])


for i in range(phi_size):
    ϕ=phi_total[i]
    for j in range(gamma_size):
        γ=gamma_total[j]
        invest_ratio[i,j]=investment_ratio(ϕ,γ,0)
```

```python
fig=plt.figure(dpi=300)
for i in range(phi_size): 
    if i==0:
        plt.plot(gamma_total,invest_ratio[0,:],label='Leontief')
    else:
        plt.plot(gamma_total,invest_ratio[i,:],label="$\sigma$={:.1f}".format(1/(1-phi_total[i])))
plt.legend()
plt.xlabel('Skill Multiplier γ')
plt.ylabel('Ratio of early to late investment')
plt.ylim(0,4)
plt.xlim(0.1,0.9)
fig.savefig('E:\研一论文\Heckman skill foramtion\文章图片\替代弹性.jpg')
plt.show()
```
